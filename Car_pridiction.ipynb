# Cleaning data

```python
#check any null values
df.isnull().sum()
```

```python
#check the duplicates
df.duplicated().value_counts()
```

```python
df.head()
```

```python
# Splitting company name from CarName column
CompanyName = df['CarName'].apply(lambda x : x.split(' ')[0])
df.insert(3,"Company car",CompanyName)
df.drop(['CarName'],axis=1,inplace=True)
df = df.rename(columns = {'Company car':'Company'})
df.head()
```

```python
df_company = pd.DataFrame(df['Company'].unique(),columns = ["company"])
df_company
```

As you see, some company have wrong name 
-"Maxda" and "Mazda"
-"Nissan" and "nissan"
-"porsche" and "porcshce"
-"toyota" and "toyouta"
-"vokswagen" and "volkswagen"

```python
#change some wrong spelling word
for company in df['Company']:
    df["Company"]=df["Company"].str.replace("maxda", "mazda")
    df["Company"]=df["Company"].str.replace("Nissan", "nissan")
    df["Company"]=df["Company"].str.replace("porcshce", "porsche")
    df["Company"]=df["Company"].str.replace("toyouta", "toyota")
    df["Company"]=df["Company"].str.replace("vokswagen", "volkswagen")
    df["Company"]=df["Company"].str.replace("vw", "volkswagen")
df["Company"].unique()
```

```python
plt.figure(figsize=(20,6))
def plot_count(x):
    df2 = pd.DataFrame(df.groupby([x])['price'].mean().sort_values(ascending = False))
    df2.plot.bar()
    plt.title(x+' vs the average price')
    plt.ylabel("The Average  price")
    plt.show()
list = ['Company','fueltype','carbody']
for name in list:
    plot_count(name)

plt.tight_layout()
```

Inference:
    1. Jagular  is brand of car  that has a average of price highest,Car prices are divided into 3 segments: high-end, mid-range and cheap
    2. Car prices use diesel still higher than gas
    3.Hardtop and convertible are high-end cars so that the price is much more expensive than others

2. carheight doesn't show any significant trend with price
